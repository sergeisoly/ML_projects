#   Исследование центральной предельной теоремы

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as sts
import math
%matplotlib inline
```

# Определяем характеристики Гамма-распределения
Про Гамма-распределение можно почитать в википедии https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D0%BC%D0%BC%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5
Плотность распредления задается следующим образом
$f\left(x\right)=x^{k-1}\cdot\frac{e^\frac{-x}\theta}{\theta^K\cdot\Gamma\left(k\right)}$, при x>0, k и $\theta$ - это коэффициенты масштаба. В scipy $\theta$ принимается за единицу.

Математическое ожидание: $E\left(x\right)=k\cdot\theta$

Дисперсия: $D\left(x\right)=k\cdot\theta^2$

```python
#посчитаем среднее и дисперсию
k=1.5  #k и  teta - коэффициенты масштаба
teta=1.
E=k*teta#среднее(мат ожидание)
D=k*teta**2.#дисперсия
```

Отсюда получаем значения:
$
E\lbrack x\rbrack=1.5\\D\lbrack x\rbrack=1.5
$

```python
#генерируем выборку объема 1000
gamma=sts.gamma(1.5)#задаем функцию
sample=gamma.rvs(1000)#генерим случайную выборку 1000 значений
x = np.linspace(0,8, 100)#определяем массив точек для построения плотности распределения
pdf = gamma.pdf(x)#вычисляем значения плотности распределения от выборки
#строим гистрограмму выборки и теоретическую функцию распределения
plt.hist(sample, bins=15, normed=True, alpha=0.5)
plt.ylabel('f(x)/number of samples')
plt.xlabel('$x$')
plt.plot(x, pdf,'r-',lw=3, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.show()
```

# Оценим распределение выборочного среднего случайной величины при разных объемах выборок
# n=5

```python
n=5
sample5=np.array([])#обозначим массив выборочных средних
x = np.linspace(0,4, 100)
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    sample5i =  gamma.rvs(n)#генерация выборки
    mean5i=sum(sample5i)/n#определение выборочного среднего (в.с.)
    sample5=np.append(sample5,mean5i)#добавляем в массив новое значение в.с.
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
E=1.5#среднее  Гамма
D=1.5#дисперсия Гамма
D5=D/n#дисперсия для нормального распределения приближающего выборку средних
sigma5=math.sqrt(D5)#сигма для нормального распределения выборки средних
norm5 = sts.norm(E, sigma5)#определяем нормальное распределение с расчетными характеристиками  Гамма
pdf5 = norm5.pdf(x)#получаем плотность распределения
#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
plt.plot(x, pdf5,label='pdf(n=5)')
plt.hist(sample5, bins =40, range=(0,4), normed=True)
plt.ylabel('f(x)/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```
